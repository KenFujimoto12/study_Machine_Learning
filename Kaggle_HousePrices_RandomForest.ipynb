```python
#欠損値　-> 'NA'　に変換
df_x.loc[df_x['GarageCond'].isnull()==True,'GarageCond']='NA'
df_x.loc[df_x['LotFrontage'].isnull()==True,'LotFrontage']='NA'
df_x.loc[df_x['MasVnrType'].isnull()==True,'MasVnrType']='NA'
df_x.loc[df_x['FireplaceQu'].isnull()==True,'FireplaceQu']='NA'
df_x.loc[df_x['BsmtQual'].isnull()==True,'BsmtQual']='NA'
df_x.loc[df_x['BsmtFinType1'].isnull()==True,'BsmtFinType1']='NA'
df_x.loc[df_x['GarageFinish'].isnull()==True,'GarageFinish']='NA'
```

```python
df_x=pd.get_dummies(df_x)
df_x.head()
```

```python
df_test=pd.read_csv('./test.csv')
df_test_x=df_test.loc[:,['Fireplaces','GarageArea','YearRemodAdd','LotArea','MSSubClass','Neighborhood',
                         'MoSold','Condition1','Condition2','OverallQual','OverallCond','ExterQual','ExterCond',
                         'TotalBsmtSF','GrLivArea','TotRmsAbvGrd','GarageCond','LotConfig','Foundation',
                         'Exterior1st','Exterior2nd','1stFlrSF','2ndFlrSF','KitchenQual','GarageCars','YearBuilt',
                         'CentralAir','MSZoning','LotFrontage','MasVnrType','BsmtQual','BsmtFinType1','BsmtFinSF1',
                         'BsmtFinSF2','FullBath','FireplaceQu','GarageFinish','WoodDeckSF','EnclosedPorch']]

df_test_x=pd.get_dummies(df_test_x)
#テストデータの欠損地の埋め方(GarageAreaの欠損地を0とした)
df_test_x.loc[df_test_x['GarageArea'].isnull(),'GarageArea']=0
df_test_x.loc[df_test_x['TotalBsmtSF'].isnull(),'TotalBsmtSF']=0
df_test_x.loc[df_test_x['GarageCars'].isnull(),'GarageCars']=0
df_test_x.loc[df_test_x['BsmtFinSF1'].isnull(),'BsmtFinSF1']=0
df_test_x.loc[df_test_x['BsmtFinSF2'].isnull(),'BsmtFinSF2']=0

#trainとsetお互いに合わないところを補う
for i in df_x.columns[df_x.columns.isin(df_test_x.columns)==False]:
    del df_x[i]
for i in df_test_x.columns[df_test_x.columns.isin(df_x.columns)==False]:
    del df_test_x[i]
```

```python
clf_RND= RandomForestRegressor()
```

```python
clf_RND.fit(df_x, df_y)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
rmse(clf_RND.predict(df_x),df_y)
```

```python
#outputに落とす
y_pred = clf_RND.predict(df_test_x)

pred_df = pd.DataFrame(y_pred, index=df_test["Id"], columns=["SalePrice"])
pred_df.to_csv('./output.csv', header=True, index_label='Id')
```
